```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as numpy
import os
```

```python
def monthly_analysis(df, value, index, func, ylabel, xlabel, opcao=None):
    if opcao == None:
        pd.pivot_table(data = df, values = value, index = index, aggfunc = func).plot(figsize=[15,5])
    elif opcao == 'sort':
        pd.pivot_table(data = df, values = value, index = index, aggfunc = func).sort_values(value).plot(figsize=[15,5])
    elif opcao == 'unstack':
        pd.pivot_table(data = df, values = value, index = index, aggfunc = func).unstack().plot(figsize=[15,5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None
    
```

```python
os.makedirs('./analises_mensais/imagens', exist_ok=True)

lista_meses = ['SINASC_RO_2019_MAR', 'SINASC_RO_2019_ABR', 'SINASC_RO_2019_MAI', 'SINASC_RO_2019_JUN', 'SINASC_RO_2019_DEZ']

for meses in lista_meses:
    df = pd.read_csv(meses+'.csv')
    monthly_analysis(df, 'IDADEMAE', 'DTNASC', 'count', 'idade da mae', 'data de nascimento')
    os.makedirs('./analises_mensais/imagens/'+ meses, exist_ok=True)
    plt.savefig('./analises_mensais/imagens/'+ meses +'/nascidos por dia.png')

for meses in lista_meses:
    df = pd.read_csv(meses+'.csv')
    monthly_analysis(df, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
    os.makedirs('./analises_mensais/imagens/'+ meses, exist_ok=True)
    plt.savefig('./analises_mensais/imagens/'+ meses +'/media peso bebe por sexo.png')

for meses in lista_meses:
    df = pd.read_csv(meses+'.csv')
    monthly_analysis(df, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
    os.makedirs('./analises_mensais/imagens/'+ meses, exist_ok=True)
    plt.savefig('./analises_mensais/imagens/'+ meses +'/media apgar1 por escolaridade mae.png')

for meses in lista_meses:
    df = pd.read_csv(meses+'.csv')
    monthly_analysis(df, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
    os.makedirs('./analises_mensais/imagens/'+ meses, exist_ok=True)
    plt.savefig('./analises_mensais/imagens/'+ meses +'/media apgar1 por escolaridade mae.png')

for meses in lista_meses:
    df = pd.read_csv(meses+'.csv')
    monthly_analysis(df, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
    os.makedirs('./analises_mensais/imagens/'+ meses, exist_ok=True)
    plt.savefig('./analises_mensais/imagens/'+ meses +'/media apgar1 por gestacao')

for meses in lista_meses:
    df = pd.read_csv(meses+'.csv')
    monthly_analysis(df, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
    os.makedirs('./analises_mensais/imagens/'+ meses, exist_ok=True)
    plt.savefig('./analises_mensais/imagens/'+ meses +'/media apgar5 por gestacao')
```
